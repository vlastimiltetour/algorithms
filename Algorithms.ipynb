## Bubble Sort
Good only for small samples of data.
Bubble sort (bublinkové řazení) je jednoduchý stabilní řadící algoritmem se složitostí O(n^2). Vylepšením bubble sortu je shakersort (oboustranný bubble sort).collection1 = [0,3,-1,3,5,3,8,45,9,-3,-9]
collection2 = ["Zoe","Adam", "Peter", "Carlos", "Jonas", "Edith"]

```python
import timeit
starttime = timeit.default_timer()


collection1 = [0,3,-1,3,5,3,8,45,9,-3,-9]
collection2 = ["Zoe","Adam", "Peter", "Carlos", "Jonas", "Edith"]


def bubbleSort(collection):
  sorted = True

  while sorted:
    sorted = False
    for i in range(len(collection)-1):
      if collection[i] < collection[i+1]:
        collection[i], collection[i+1] = collection[i+1], collection[i]
        sorted = True

  return collection


print(bubbleSort(collection1))
print(bubbleSort(collection2)) 
print("Duration:", timeit.default_timer() - starttime)
```

## Shaker Sort 🍹

```python
#Shaker Sort v.2
array1 = [0,3,-1,3,5,3,8,45,9,-3,-9]
array2 = ["Zoe","Adam", "Peter", "Carlos", "Jonas", "Edith"]

def Shaker_Sort(a):
  sorted = True

  while sorted:
    sorted = False

    for i in range(len(a)-1):
      if a[i] > a[i+1]:
        a[i], a[i+1] = a[i+1], a[i]
        sorted = True

    for i in range(-1,0,-1):
      if a[i] > a[a-1]:
        a[i], a[i-1] = a[i-1], a[i]
        sorted = True

  return a

print(Shaker_Sort(array1))
```

## Selection Sort
Selection sort is a simple sorting algorithm with asymptotic complexity O(n^{2}). In comparison with other quadratic sorting algorithms it almost always outperforms bubble sort, but it is usually slower than insertion sort. The advantage of selection sort over algorithms with O(n \\cdot \\log{n}) (quicksort, heapsort, merge sort) asymptotic complexity is it's constant memory complexity.



```python
array1 = [0,3,-1,3,5,3,8,45,9,-3,-9]
array2 = ["Zoe","Adam", "Peter", "Carlos", "Jonas", "Edith"]

def Selection_Sort(a):
  for i in range(len(a)-1):
    #print(i)
    #usr = input()
    min_index = i # i = iterator in the list (a), it is the >>order<<

    for j in range(i+1,len(a)): #j is a second iterator in the list (a), i.e., from i+1
      if a[min_index] > a[j]: # if the zero value, e.g. 0 is higher than the first value of 3, 
        #print(a[min_index])
        #user = input()
        min_index = j
    
    a[i],a[min_index] = a[min_index], a[i]
  
  return a

print(Selection_Sort(array1))
```

## Quick Sort
Quicksort je velmi rychlý nestabilní řadící algoritmus na principu Divide et Impera (rozděl a panuj) s asymptotickou složitostí O(n^{2}) a s očekávanou složitostí O(n \\cdot \\log{n}). Algoritmus vymyslel v roce 1962 Sir Charles Antony Richard Hoare.

```python
collection1 = [0,3,-1,3,5,3,8,45,9,-3,-9]
collection2 = ["Zoe","Adam", "Peter", "Carlos", "Jonas", "Edith"]

def quickSort(collection):
  if len(collection) < 1:
    return collection

  pivot = collection[0]
  lower = [x for x in collection[1:] if x <= pivot]
  higher = [x for x in collection[1:] if x > pivot]

  return quickSort(lower) + [pivot] + quickSort(higher)

print(quickSort(collection1))
print(quickSort(collection2))
```

## Merge Sort

```python
from collections import Counter
starttime = timeit.default_timer()

array1 = [0,3,-1,3,5,3,8,45,9,-3,-9]
array2 = ["Zoe","Adam", "Peter", "Carlos", "Jonas", "Edith"]

def MergeSort(a):
  if len(a) > 1:
    mid = len(a) // 2
    left = a[mid:]
    right = a[:mid]

    i = j = k = 0

    MergeSort(left)
    MergeSort(right)

    while i < len(left) and j < len(right):
      if left[i] <= right[j]:
        a[k] = left[i]
        i += 1
      else:
        a[k] = right[j]
        j += 1
      
      k += 1
    
    while i < len(left):
      a[k] = left[i]
      i += 1
      k += 1
    
    while j < len(right):
      a[k] = right[j]
      j += 1
      k += 1
    
    return a
  
print(MergeSort(array1))
```

```python
##posledni testovaci verze proc to kurwa nejde

import timeit
starttime = timeit.default_timer()

array1 = [0,3,-1,3,5,3,8,45,9,-3,-9]
array2 = ["Zoe","Adam", "Peter", "Carlos", "Jonas", "Edith"]

def merge_sort(A):
  #until the length of the not higher than 1
  if len(A) > 1:
    middle = len(A) // 2
    left = A[:middle]
    right = A[middle:]

    #The list is divided into left and right in each recursive call 
    #until two adjacent elements are obtained.


    merge_sort(left)
    merge_sort(right)

    i = 0 # traverses the left, i.e. left half
    j = 0 # traverses the j, i. e. right half
    k = 0 # traverses whole lists

    #Now begins the sorting process. The i and j iterators 
    #traverse the two halves in each call. 
    #The k iterator traverses the whole lists and makes changes along the way.


    while i < len(left) and j < len(right):
      if left[i] < right[j]:
        A[k] = left[i]
        i += 1
      else:
        A[k] = right[j]
        j += 1

      k += 1
    
    while i < len(left):
      A[k] = left[i]
      i += 1
      k += 1
    
    while j < len(right):
      A[k] = right[j]
      j += 1
      k += 1

    return A

print(merge_sort(array1))
print(merge_sort(array2))
print("Duration:", timeit.default_timer() - starttime)
```
