Here's a table of some common algorithms with their major characteristics:

| Algorithm      | Type                   | Best Case Time Complexity | Worst Case Time Complexity | Space Complexity | Stable | In-Place | Notes                                     |
|----------------|------------------------|---------------------------|----------------------------|-------------------|--------|----------|-------------------------------------------|
| Quick Sort     | Divide and Conquer     | O(n log n)                | O(n²)                      | O(log n)          | No     | Yes      | Efficient for large datasets, but worst case can be mitigated with randomization. |
| Merge Sort     | Divide and Conquer     | O(n log n)                | O(n log n)                 | O(n)              | Yes    | No       | Stable sort, good for linked lists and large datasets.                           |
| Bubble Sort    | Comparison-based       | O(n)                      | O(n²)                      | O(1)              | Yes    | Yes      | Simple but inefficient, mainly used for educational purposes.                   |
| Insertion Sort | Comparison-based       | O(n)                      | O(n²)                      | O(1)              | Yes    | Yes      | Efficient for small datasets and nearly sorted data.                            |
| Selection Sort | Comparison-based       | O(n²)                     | O(n²)                      | O(1)              | No     | Yes      | Simple but inefficient, useful when memory write is costly.                     |
| Heap Sort      | Comparison-based       | O(n log n)                | O(n log n)                 | O(1)              | No     | Yes      | Efficient and in-place, but not stable.                                         |
| Counting Sort  | Non-comparison-based   | O(n + k)                  | O(n + k)                   | O(k)              | Yes    | No       | Efficient for small range of integers, k is the range of input values.          |
| Radix Sort     | Non-comparison-based   | O(nk)                     | O(nk)                      | O(n + k)          | Yes    | No       | Efficient for large numbers of small integers, k is the number of digits.       |
| Shell Sort     | Comparison-based       | O(n log n)                | O(n²)                      | O(1)              | No     | Yes      | Generalization of insertion sort, efficient for medium-sized datasets.          |

### Explanation of Terms:

- **Type**: The fundamental approach the algorithm uses to sort the data.
- **Best Case Time Complexity**: The time complexity of the algorithm in the best-case scenario.
- **Worst Case Time Complexity**: The time complexity of the algorithm in the worst-case scenario.
- **Space Complexity**: The amount of extra space the algorithm requires.
- **Stable**: Whether the algorithm maintains the relative order of records with equal keys.
- **In-Place**: Whether the algorithm requires only a constant amount of additional memory space.


```python
def quicksort(array):

    #block endless recursion
    if len(array) < 2:
        #return sorted, finished array
        return array

    #if the array needs sorting
    else:
        #decide on the pivot
        pivot = len(array) // 2


        #break array to 3 pieces and find pivot
        #iterate and compare against pivot
        left = [x for x in array if x < array[pivot]]
        mid = [x for x in array if x == array[pivot]]
        right = [x for x in array if x > array[pivot]]

    #one loop is finished, need to go through it again before it's sorted
    return quicksort(left) + mid + quicksort(right)


print(quicksort([2,6,1,9,8,6]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for time complexity
n = np.linspace(1, 1000, 100)
best_case = n * np.log2(n)
average_case = n * np.log2(n)
worst_case = n**2

# Plotting time complexity
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.plot(n, best_case, label="Best Case: O(n log n)", color="green")
plt.plot(n, average_case, label="Average Case: O(n log n)", color="blue")
plt.plot(n, worst_case, label="Worst Case: O(n^2)", color="red")
plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.title('Quick Sort Time Complexity')
plt.legend()

# Data for space complexity
best_space = np.log2(n)
worst_space = n

# Plotting space complexity
plt.subplot(1, 2, 2)
plt.plot(n, best_space, label="Best/Average Case: O(log n)", color="blue")
plt.plot(n, worst_space, label="Worst Case: O(n)", color="red")
plt.xlabel('Input Size (n)')
plt.ylabel('Space (recursion depth)')
plt.title('Quick Sort Space Complexity')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Bubble sort
'''Bubble Sort is the simplest sorting algorithm that works by repeatedly swapping the adjacent elements if they are in the wrong order.
This algorithm is not suitable for large data sets as its average and worst-case time complexity is quite high.

Bubble Sort Algorithm
In Bubble Sort algorithm,

- traverse from left and compare adjacent elements and the higher one is placed at right side.
- In this way, the largest element is moved to the rightmost end at first.
- This process is then continued to find the second largest and place it and so on until the data is sorted.'''

def bubble_sort(arr):
    for i in range(len(arr)):
      for j in range(len(arr)- i - 1):
        if arr[j] > arr[j+1]:
          arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print(bubble_sort([2,6,1,9,8,6]))
```
